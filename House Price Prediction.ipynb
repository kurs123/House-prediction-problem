## Feature engineering

```python
sns.distplot(data.sqft_living)
```

```python
sns.distplot(data.sqft_living15)
```

```python
#standardization doesnt reduce the skewness but it compresses the spread of the variable.
zzz=data.sqft_living
z=(zzz-np.mean(zzz))/(np.std(zzz))
sns.distplot(z)
```

```python
zzz=data2.sqft_living15
z=(zzz-np.mean(zzz))/(np.std(zzz))
sns.distplot(z)
```

```python
#logarithmic transformation reduces the spread and also the skewness.
ggg=data.sqft_living
g=np.log(ggg)
sns.distplot(g)
```

```python
ggg=data2.sqft_living15
g=np.log(ggg)
sns.distplot(g)
```

```python
data['log(sqft_living)']=np.log(data.sqft_living)
data['log(sqft_living15)']=np.log(data.sqft_living15)
data['sqrt(sqft_living)']=((data.sqft_living)-(np.mean(data.sqft_living)))/np.std(data.sqft_living)
data['sqrt(sqft_living15)']=((data.sqft_living15)-(np.mean(data.sqft_living15)))/np.std(data.sqft_living15)
data['log(sqft_above)']=np.log(data.sqft_above)
data['sqrt(sqft_basement)']=np.sqrt(data.sqft_basement)
data.transpose()
```

```python
# Slicing date to get just the year

for i in data.date:
    i=i[:4]
    data.date=i
data
```

```python
#data.date is in in the object format so we convert it into float 

data.date=data.date.astype(float)

data['age']=data['date']-data['yr_built']
data=data.drop(['date','yr_built'],axis=1)
data
```

## Model fit part 3

## Backward elimination

#### Step 1:​
#### Considering all the attributes:​
#### X= ['bedrooms', 'bathrooms', 'floors’, log(sqft_living),sqft_lot,'waterfront’,   'view','condition', 'grade', ‘log(sqft_above)', ‘sqrt(sqft_basement)’,    'lat’,'long’,'log(sqft_living15)’, sqft_lot15,'age’]​

#### Step 2:​
#### Dropping the columns with p value greater than 0.05.​
#### The predictors ‘sqft_lot15’ and ’floors’ show p>0.05 hence they are removed and step 1 is repeated. This is done until all the predictors with p>0.05 are eliminated.​

#### Step 3:​
#### Considering the remaining variables after all predictors with p>0.05 are eliminated .​
#### X_final='bedrooms', 'bathrooms', 'log(sqft_living)', 'sqft_lot', 'waterfront', 'view', 'condition', 'grade', 'log(sqft_above)', 'sqrt(sqft_basement)', 'lat', 'long', 'log(sqft_living15)', 'age'] ​

## Forward selection

#### Using mlxtend library :​

#### from mlxtend.feature_selection import SequentialFeatureSelector as sfs​

#### clf=LinearRegression().fit(X,y)​

#### sfs1 = sfs(clf,k_features=6,forward=True, floating=False,​

#### verbose=2,scoring='r2',cv=5)​

#### This gives us a set of 6 predictors which are the best ranking predictors for our regression based on the scoring R2.​

#### The predictors are:​

#### X=['bathrooms’, 'waterfront’ , 'view’ ,’grade’,  'lat’, 'age']​

## LINEAR REGRESSION ON TEST DATA

```python
X=data_test[['bedrooms', 'bathrooms', 'sqft_living', 'floors', 'waterfront', 'view',
       'condition', 'grade', 'sqft_above', 'sqft_basement', 'lat', 'long',
       'sqft_living15', 'sqft_lot15', 'log(sqft_living)', 'age',
       'sqrt(sqft_living)', 'sqrt(sqft_living15)']]
y=data_test.price

model=LinearRegression().fit(X,y)

yhat=model.predict(X)
error=y-yhat

SSE=np.sum(np.square(error))

MSE=SSE/len(data_test)

RMSE=np.sqrt(MSE)

print(f"RMSE={RMSE}")
print(f"R2={model.score(X,y)}")
```
